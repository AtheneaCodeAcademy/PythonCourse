## üìå ¬øQu√© son las estructuras de datos?
Las **estructuras de datos** son elementos l√≥gicos y organizativos utilizados para **almacenar**, **gestionar** y **acceder eficientemente** a los datos. Permiten representar y manipular colecciones de informaci√≥n de manera que se optimicen operaciones como la **inserci√≥n**, **eliminaci√≥n**, **b√∫squeda** y **recorrido**.

Una estructura de datos define no solo el formato de almacenamiento, sino tambi√©n el **conjunto de operaciones v√°lidas** que se pueden realizar sobre los datos, as√≠ como las **relaciones l√≥gicas** entre ellos.

`Ejemplos`

```python
lista = [elemento1, elemento2, elemento3]
```

```python
tupla = (elemento1, elemento2)
```

```python
conjunto = {elemento1, elemento2, elemento3, elemento4}
```

```python
diccionario = {clave1:valor1, clave2:valor2, clave3:valor3, clave4:valor4}
```

## üß™ Tipos de estructuras
### `1) Listas`

Las **listas** (list) es una estructura de datos **lineal**, **ordenada** y **mutable** que permite almacenar una colecci√≥n de elementos heterogeneos (de distintos tipos), accesibles por su **√≠ndice**.

Se consideran **arreglos din√°micos** (dynamic arrays) lo que significa que pueden **crecer** o **reducir** su tama√±o en tiempo de ejecuci√≥n.

`Caracter√≠sticas`
- Ordenadas: mantienen el orden en que se insertan los elementos.
- Indexadas: cada elemento tiene una posici√≥n asociada (√≠ndice), comenzando desde el valor 0.
- Mutables: pueden modificarse despu√©s de su creaci√≥n (agregar, eliminar o cambiar elementos).
- Admiten duplicados: los elementos no necesitan ser √∫nicos.
- Heterog√©neas: pueden contener elementos de diferentes tipos (n√∫meros, cadenas, listas, etc.).

````python
# Ejemplo 1
lista_de_numeros = [1, 2, 3, 4, 5]
````

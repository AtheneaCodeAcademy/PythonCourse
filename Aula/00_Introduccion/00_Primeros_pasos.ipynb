```python
# Gr√°ficos... y algo de matem√°ticas (si quieres)
import numpy as np
import matplotlib.pyplot as plt

def sin_vs_cos(n_vueltas):
    grados = np.linspace(0, n_vueltas * 360, num=1000)
    radianes = np.deg2rad(grados)

    seno = np.sin(radianes)
    coseno = np.cos(radianes)

    plt.figure(figsize=(10, 5))
    plt.plot(grados, seno, label='Seno', color='blue')
    plt.plot(grados, coseno, label='Coseno', color='red')
    plt.title(f' sin(x) vs. cos(x)')
    plt.xlabel('Grados')
    plt.ylabel('y')
    plt.grid(False)
    plt.legend()
    plt.tight_layout()
    plt.show()

sin_vs_cos(2)
```

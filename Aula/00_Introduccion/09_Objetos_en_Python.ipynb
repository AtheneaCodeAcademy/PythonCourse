## ¿Qué es un objeto?
Python es un lenguaje **orientado a objetos***, lo que significa que **todo** en Python es un **objeto**.
Un **objeto** es una estructura en memória que contiene:
- **Tipo** (`int`, `str`, `list`, `dict` y un infinito etcétera...)
- **Valor** (contenido del objeto)
- **Atributos** y **métodos** (carácterísticas y funciones asociadas al objeto)
- **Identificador** (dirección de memória)

Decir que en **Python** todo es un objeto significa que todos los elementos en Python, **sin importar su tipo**, son **instancias** (objetos) de una clase. Esto incluye números, cadenas de texto, listas, funciones, clases, módulos, etc.

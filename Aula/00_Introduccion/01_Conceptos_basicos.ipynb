## üìö Tipos de datos en Python
En Python, cada **valor** pertenece a un **tipo de dato**, que determina el tipo de informaci√≥n que representa y las operaciones que se pueden realizar sobre √©l.

- `Primitivos`

| Tipo         | Ejemplos        | Descripci√≥n                                             |
| ------------ | --------------- | -------------------------------------------             |
| **int**      | `1`, `-5`       | N√∫meros enteros (positivos, negativos y sin decimales)  |
| **float**    | `3.14`, `-10.1` | N√∫meros decimales                                       |
